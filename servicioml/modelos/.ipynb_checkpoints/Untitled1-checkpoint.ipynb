```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import string
from string import punctuation
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#Read data de neustro local

df =  pd.read_csv("C:/Users/hjjc1/Desktop/tripadvisor_hotel_reviews.csv")


df.head()
```

```python
#vamos a aplicar la regla de rting >3 positivo =3 neutro   <3 negativo

df['Sentiment'] = df['Rating'].apply(lambda x: 'Positive' if x>3 else('Neutral' if x==3 else 'Negative'))

df.head(10)
```

```python
#vizualizacion de los datos
sns.set_style('whitegrid')
sns.countplot(x='Rating',data=df, palette='YlGnBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Sentiment',data=df, palette='summer')
```

```python
#Vamos a procesar los datos que tenemos
df_Positive = df[df['Sentiment'] == 'Positive']
df_Neutral = df[df['Sentiment'] == 'Neutral']
df_Negative = df[df['Sentiment'] == 'Negative']

df_Negative
```

```python
#rellenamos los datos con verdadero hasta llegar a los 8000
df_Neutral_over = df_Neutral.sample(8000, replace=True)
df_Negative_over = df_Negative.sample(8000, replace=True)
df = pd.concat([df_Positive, df_Neutral_over, df_Negative_over], axis=0)

df_Negative_over
```

```python
def get_text_processing(text):
    stpword = stopwords.words('english')
    no_punctuation = [char for char in text if char not in string.punctuation]
    no_punctuation = ''.join(no_punctuation)
    return ' '.join([word for word in no_punctuation.split() if word.lower() not in stpword])
```

```python
df['review'] = df['Review'].apply(get_text_processing)
df.head(10)
```

```python
df_final = df[['review', 'Sentiment']]
df_final.head()
```

```python
one_hot = pd.get_dummies(df["Sentiment"])
df_final.drop(["Sentiment"], axis=1, inplace=True)
df_final = pd.concat([df_final, one_hot], axis=1)
df_final.head()
```

```python
# vamos a entrenar nuestro modelo y para eso separamos lo que vamos a usar para separar lo que usamos para entrenar y testear

X = df_final["review"].values
y = df_final.drop("review", axis=1).values
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)
```

```python
#aplicamos la vectorizacion .
vect = CountVectorizer()
X_train = vect.fit_transform(X_train)
X_test = vect.transform(X_test)
```

```python
#aplicamos frecuencia
tfidf = TfidfTransformer()
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
X_train = X_train.toarray()
X_test = X_test.toarray()
```

```python
#Agregamos diferentes capas al modelo 
model = Sequential()
model.add(Dense(units=12673, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=4000, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=500, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=3, activation="softmax"))
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
early_stop = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=2)
```

```python
#Ajustamos nuestro modelo para que se adapte
model.fit ( 
    x = X_train, 
    y = y_train, 
    batch_size = 256 , epochs = 
    100, 
    validation_data = (X_test, y_test), 
    verbose = 1, 
    callbacks = early_stop, 
)
```

```python
#Evaluacion del modelo 
model_score = model.evaluate(X_test, y_test, batch_size=64, verbose=1)
print("Test accuracy:", model_score[1])
```

```python
#Predicion de resultado 
preds = model.predict(X_test)
preds
```
